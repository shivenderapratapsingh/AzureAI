```python
# Scenario
# You’re training healthcare analysts.
# You want them to see how to connect securely to Azure OpenAI, ask questions about AI in healthcare, 
# and then deploy the interaction into a Gradio interface so they can try it themselves.
```

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr

# load environment variables
load_dotenv()

# create Azure OpenAI client
client = OpenAI(
    base_url=os.getenv("endpoint"),
    api_key=os.getenv("api_key")
)

model_name = os.getenv("deployment_name")


# function for healthcare Q&A
def healthcare_ai_assistant(question):

    response = client.responses.create(
        model=model_name,
        input=f"""
You are an AI assistant helping healthcare analysts understand AI concepts safely and ethically.

Question: {question}

Provide clear explanation with healthcare example.
"""
    )

    return response.output_text


# Gradio interface
interface = gr.Interface(
    fn=healthcare_ai_assistant,
    inputs=gr.Textbox(
        placeholder="Ask about AI in healthcare...",
        label="Healthcare AI Question"
    ),
    outputs=gr.Textbox(label="AI Response"),
    title="Healthcare AI Training Assistant",
    description="Ask questions about AI applications in healthcare"
)

# run app
interface.launch()
```

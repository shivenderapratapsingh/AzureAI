```python
data = [
    "Transaction successful. Ref ID: TXN#A12345",
    "Your payment with TXN # B67890 has been processed",
    "Failed transaction TXN#C54321 due to insufficient balance"
]
ans=[]
for val in data:
    if (re.search("TXN#[A-Z]{1}[0-9]+",val)):
        ans.extend(re.findall("TXN#[A-Z]{1}[0-9]+",val))

print(ans)

# re.findall(r"(?=.[A-Z])")
```

```python
from openai import OpenAI
from azure.storage.blob import BlobServiceClient

# AI model
client = OpenAI(
    api_key="",
    base_url="https://newfoundaryzoro.openai.azure.com/openai/v1"
)

paper_text = open("paper.txt").read()

response = client.chat.completions.create(
    model="gpt-4.1",
    messages=[
        {"role": "system", "content": "Summarize research paper"},
        {"role": "user", "content": paper_text}
    ]
)

summary = response.choices[0].message.content

print(summary)

```

```python
# storage
conn_str =""


blob_service_client = BlobServiceClient.from_connection_string(conn_str)

container_name = "shivcontainer"

container_client = blob_service_client.get_container_client(container_name)

try:
    container_client.create_container()
except:
    pass

blob_client = blob_service_client.get_blob_client(
    container=container_name,
    blob="summary.txt"
)

blob_client.upload_blob(summary, overwrite=True)

print("✅ Uploaded to Azure Storage")
```

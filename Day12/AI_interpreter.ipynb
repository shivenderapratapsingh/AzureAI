```python
from openai import OpenAI

client = OpenAI(
    api_key="APIKEYHERE",
    base_url="https://newfoundaryzoro.openai.azure.com/openai/v1"
)

report_text = open("medicalreport.txt").read()

response = client.chat.completions.create(
    model="gpt-4.1",
    messages=[
        {
            "role": "system",
            "content": "Explain medical reports in simple language"
        },
        {
            "role": "user",
            "content": report_text
        }
    ]
)

interpretation = response.choices[0].message.content

print(interpretation)
```

```python
from azure.storage.blob import BlobServiceClient

conn_str=conn_str = ""

blob_service_client = BlobServiceClient.from_connection_string(conn_str)

container_name="shivcontainer"

container_client = blob_service_client.get_container_client(container_name)

try:
    container_client.create_container()
except:
    pass

blob_client = blob_service_client.get_blob_client(
    container=container_name,
    blob="report_result.txt"
)

blob_client.upload_blob(interpretation, overwrite=True)

print("✅ Report stored in cloud")
```

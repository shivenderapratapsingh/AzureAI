```python
from openai import OpenAI
from dotenv import load_dotenv
import os
load_dotenv()

client = OpenAI(
    base_url="https://aifoundarycreate-1223.openai.azure.com/openai/v1/",
    api_key=os.getenv("apikey")
)

def interview(mode):

    system_prompt = f"""
You are an AI Interviewer.

Interview Mode: {mode}

Your job:
1. Ask one interview question
2. Evaluate candidate answer
3. Give feedback
4. Suggest improvement

Keep response structured:
Question:
Feedback:
Improvement Tip:
"""

    while True:

        response = client.chat.completions.create(
            model="gpt-4.1",
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": "Ask next question"}
            ]
        )

        question = response.choices[0].message.content

        print("\n", question)

        answer = input("\nYour Answer: ")

        if answer.lower() == "exit":
            break

        feedback = client.chat.completions.create(
            model="gpt-4.1",
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": f"Question: {question}\nCandidate Answer: {answer}"}
            ]
        )

        print("\nAI Feedback:\n")
        print(feedback.choices[0].message.content)


print("Select mode:")
print("1. HR")
print("2. Technical")
print("3. Stress")

choice = input("Enter choice: ")

mode_map = {
    "1": "HR interview",
    "2": "Technical interview",
    "3": "Stress interview"
}

interview(mode_map[choice])
```

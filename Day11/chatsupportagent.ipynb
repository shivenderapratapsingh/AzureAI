```python
from openai import OpenAI
from langchain_openai import OpenAI
from dotenv import load_dotenv
import os
load_dotenv()
endpoint = "https://aifoundarycreate-1223.openai.azure.com/openai/v1/"
deployment_name = "gpt-4.1"

client = OpenAI(
    base_url=endpoint,
    api_key=os.getenv("apikey")
)

def ask_support(question):

    response = client.chat.completions.create(
        model="gpt-4.1",   # deployment name
        messages=[
            {
                "role": "system",
                "content": "You are a customer support assistant. Answer politely and help users with queries about orders, refunds, payments, accounts, and product issues."
            },
            {
                "role": "user",
                "content": question
            }
        ]
    )

    return response.choices[0].message.content


while True:

    user_input = input("Customer: ")

    if user_input.lower() == "exit":
        break

    print("Bot:", ask_support(user_input))
```
